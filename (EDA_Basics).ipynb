Q1. What is the average, median, and standard deviation of house prices?

```python
import pandas as pd

# Replace this list with your actual house price data
house_prices = [350000, 420000, 290000, 850000, 410000, 320000, 550000]

# Create a pandas Series
prices_series = pd.Series(house_prices)

# Calculate metrics
print(f"Average (Mean): ${prices_series.mean():,.2f}")
print(f"Median:         ${prices_series.median():,.2f}")
print(f"Std Deviation:  ${prices_series.std():,.2f}")
```

Q3. Check if the price data is skewed.

```python
import pandas as pd

# Your house price data
house_prices = [350000, 420000, 290000, 850000, 410000, 320000, 550000]
prices_series = pd.Series(house_prices)

# Calculate skewness
skewness = prices_series.skew()

print(f"Skewness Coefficient: {skewness:.2f}")

# Interpret the result
if skewness > 0.5:
    print("The data is Positively Skewed (Right-skewed) - pulled by high prices.")
elif skewness < -0.5:
    print("The data is Negatively Skewed (Left-skewed) - pulled by low prices.")
else:
    print("The data is relatively Symmetric.")
```

Q4. What is the average price for each number of bedrooms?

```python
import pandas as pd

# Combined data from your previous steps
data = {
    'price': [350000, 420000, 290000, 850000, 410000, 320000, 550000],
    'bedrooms': [3, 4, 3, 5, 3, 4, 4]
}

df = pd.DataFrame(data)

# Group by bedrooms and calculate the average price
avg_price_by_bedroom = df.groupby('bedrooms')['price'].mean().reset_index()

# Format the output for readability
avg_price_by_bedroom['price'] = avg_price_by_bedroom['price'].map('${:,.2f}'.format)
print(avg_price_by_bedroom.to_string(index=False))
```

Q5.  What is the relationship between living area and price? (Use : Area without Basement)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Update your dataset to include the living area (sq ft) for each house
data = {
    'price': [350000, 420000, 290000, 850000, 410000, 320000, 550000],
    'bedrooms': [3, 4, 3, 5, 3, 4, 4],
    'area_no_basement': [1500, 1800, 1200, 3500, 1700, 1400, 2400] # Example area data in sq ft
}

df = pd.DataFrame(data)

# 2. Calculate the Pearson correlation coefficient
correlation = df['area_no_basement'].corr(df['price'])
print(f"Correlation between Living Area and Price: {correlation:.2f}")

# 3. Interpret the relationship
if correlation > 0.7:
    print("There is a strong positive relationship (as area increases, price increases significantly).")
elif correlation > 0.3:
    print("There is a moderate positive relationship.")
else:
    print("There is a weak or no linear relationship.")

# 4. Create a quick scatterplot to visualize it
plt.figure(figsize=(6, 4))
sns.regplot(data=df, x='area_no_basement', y='price', scatter_kws={'s':100}, line_kws={'color':'red'})
plt.title('Living Area vs. House Price')
plt.xlabel('Living Area (Without Basement) sq ft')
plt.ylabel('Price ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Q6. Identify any anomalies where houses have high prices but low areas.

```python
import pandas as pd

# Data from your previous steps
data = {
    'price': [350000, 420000, 290000, 850000, 410000, 320000, 550000],
    'bedrooms': [3, 4, 3, 5, 3, 4, 4],
    'area_no_basement': [1500, 1800, 1200, 3500, 1700, 1400, 2400]
}

df = pd.DataFrame(data)

# 1. Calculate price per square foot
df['price_per_sqft'] = df['price'] / df['area_no_basement']

# 2. Define an anomaly threshold (e.g., more than 1.5 standard deviations above the average ratio)
mean_ratio = df['price_per_sqft'].mean()
std_ratio = df['price_per_sqft'].std()
threshold = mean_ratio + (1.5 * std_ratio)

anomalies = df[df['price_per_sqft'] > threshold]

# 3. Print results
print(f"Cutoff threshold for high price/low area anomaly: ${threshold:.2f} per sq ft\n")
if anomalies.empty:
    print("No anomalies found! All house prices scale proportionally with their living area.")
else:
    print("Anomalies detected (High price, low area):")
    print(anomalies[['price', 'area_no_basement', 'price_per_sqft']])
```

Q7. Compare average price based on number of floors and houses with or without waterfront.

```python
import pandas as pd

# 1. Update the dataset with example values for 'floors' and 'waterfront'
# (Make sure these match your actual dataset rows)
data = {
    'price': [350000, 420000, 290000, 850000, 410000, 320000, 550000],
    'bedrooms': [3, 4, 3, 5, 3, 4, 4],
    'area_no_basement': [1500, 1800, 1200, 3500, 1700, 1400, 2400],
    'floors': [1, 2, 1, 2, 1, 1, 2],                 # Number of floors
    'waterfront': [0, 0, 0, 1, 0, 0, 1]              # 0 = No Waterfront, 1 = Has Waterfront
}

df = pd.DataFrame(data)

# 2. Map waterfront values to readable labels for the final table
df['waterfront'] = df['waterfront'].map({1: 'Waterfront', 0: 'No Waterfront'})

# 3. Create a pivot table showing the average price
pivot_table = df.pivot_table(
    values='price',
    index='floors',
    columns='waterfront',
    aggfunc='mean'
)

# 4. Format numbers to currency style
formatted_pivot = pivot_table.applymap(lambda x: f"${x:,.2f}" if pd.notnull(x) else "N/A")

print("--- Average Price by Floors and Waterfront Views ---")
print(formatted_pivot)
```

Q9. Which location (zipcode/area) has the highest average price?

```python
import pandas as pd

# 1. Update the dataset to include a 'zipcode' or 'location' column
# (Make sure these match your actual dataset rows)
data = {
    'price': [350000, 420000, 290000, 850000, 410000, 320000, 550000],
    'bedrooms': [3, 4, 3, 5, 3, 4, 4],
    'area_no_basement': [1500, 1800, 1200, 3500, 1700, 1400, 2400],
    'floors': [1, 2, 1, 2, 1, 1, 2],
    'waterfront': [0, 0, 0, 1, 0, 0, 1],
    'zipcode': ['98001', '98002', '98001', '98004', '98002', '98001', '98004'] # Example Zipcodes
}

df = pd.DataFrame(data)

# 2. Group by zipcode and calculate the average price
avg_price_by_zip = df.groupby('zipcode')['price'].mean().reset_index()

# 3. Sort to find the highest average price first
avg_price_by_zip = avg_price_by_zip.sort_values(by='price', ascending=False)

# 4. Format the numbers for clean reading
avg_price_by_zip['price'] = avg_price_by_zip['price'].map('${:,.2f}'.format)

print("--- Average House Price by Zipcode ---")
print(avg_price_by_zip.to_string(index=False))

print(f"\nThe location with the highest average price is Zipcode: {avg_price_by_zip.iloc[0]['zipcode']}")
```
