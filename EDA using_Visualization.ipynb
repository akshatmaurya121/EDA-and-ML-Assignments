Q1. Create a countplot to show the number of passengers by gender

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Set the visualization style
sns.set_theme(style="whitegrid")

# Create the countplot for gender (Sex)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Sex", palette="pastel")

# Add titles and labels
plt.title("Number of Passengers by Gender", fontsize=14)
plt.xlabel("Gender (Sex)", fontsize=12)
plt.ylabel("Passenger Count", fontsize=12)

# Display the plot
plt.show()
```

Q8. Create a heatmap to show correlation between numerical variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Select only numerical columns to avoid errors in calculation
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Set the visualization style
sns.set_theme(style="white")

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)

# Add title
plt.title("Correlation Heatmap of Numerical Variables", fontsize=14)

# Display the plot
plt.show()
```

Q9. Identify which passenger class paid the highest fare on average.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Titanic.csv")

# Calculate the mean fare grouped by Passenger Class (Pclass)
average_fares = df.groupby("Pclass")["Fare"].mean()

print("Average fare paid by each passenger class:")
print(average_fares)

# Programmatically find the class with the highest average fare
highest_fare_class = average_fares.idxmax()
highest_fare_value = average_fares.max()

print(f"\nPassenger Class {highest_fare_class} paid the highest fare on average (£{highest_fare_value:.2f}).")
```
